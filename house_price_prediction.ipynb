```python
X=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['beds']), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Instantiate StandardScaler to normalize data
scaler = StandardScaler()

# Fit and transform your training data
X_train_scaled = scaler.fit_transform(X_train)

# Instantiate LinearRegression without the 'normalize' parameter
lr = LinearRegression()

# Fit the model with scaled data
lr.fit(X_train_scaled, y_train)

# Transform your test data before prediction
X_test_scaled = scaler.transform(X_test)

# Now you can make predictions
predictions = lr.predict(X_test_scaled)
```

```python
pipe = make_pipeline(column_trans,scaler, lr)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

```python
ridge = Ridge()
pipe = make_pipeline(column_trans,scaler, ridge)
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```
